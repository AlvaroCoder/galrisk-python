```python
def filtrar_mayores_impactos_top(resultados, top_n=6):
    """
    Filtra y ordena los resultados según la variación absoluta, de mayor a menor.

    Args:
        resultados (list): Lista de diccionarios con las variaciones calculadas.
        top_n (int): Número de resultados a retornar (por defecto 6).

    Returns:
        list: Lista de los 'top_n' elementos con mayor variación absoluta.
    """

    if not resultados:
        return []

    # Filtrar resultados válidos con variación absoluta definida
    resultados_filtrados = [
        r for r in resultados
        if r.get("Variación Absoluta") is not None
    ]
    # Ordenar de mayor a menor según variación absoluta
    resultados_ordenados = sorted(
        resultados_filtrados,
        key=lambda x: x["Variación Absoluta"],
        reverse=True
    )

    # Retornar solo los primeros 'top_n'
    return resultados_ordenados[:top_n]
```

```python
import numpy as np
import pandas as pd
from scipy.stats import triang, beta

def generar_tablas_aleatorias_top(top_variables, n=1000):
    """
    Genera tablas de valores aleatorios para las variables top más influyentes.

    Args:
        top_variables (list): Lista de variables top con 'Celda Raíz' y 'Valor Original'.
        n (int): Número de muestras aleatorias a generar para cada variable.

    Returns:
        dict: Diccionario con DataFrames para cada variable.
    """
    tablas = {}
    distribuciones = {}
    for var in top_variables:
        nombre = f"{var['Hoja Raíz']}!{var['Celda Raíz']}"
        moda = var["Valor Original"]

        print(f"\nConfiguración para variable: {nombre} (Moda: {moda})")
        minimo = float(input(f"→ Ingrese el valor mínimo de {nombre} -> {moda}: "))
        maximo = float(input(f"→ Ingrese el valor máximo de {nombre} -> {moda}: "))

        while not (minimo <= moda <= maximo):
            print("⚠️  La moda debe estar entre mínimo y máximo.")
            minimo = float(input("→ Ingrese el valor mínimo: "))
            maximo = float(input("→ Ingrese el valor máximo: "))

        distribucion = input("→ Tipo de distribución (PERT / Normal / Triangular): ").strip().lower()
        while distribucion not in ["pert", "normal", "triangular"]:
            distribucion = input("⚠️  Distribución inválida. Ingrese (PERT / Normal / Triangular): ").strip().lower()

        # Generar muestras
        if distribucion == "normal":
            mu = moda
            sigma = (maximo - minimo) / 6  # Aprox. 99.7% de valores en rango
            muestras = np.random.normal(mu, sigma, n)
            muestras = np.clip(muestras, minimo, maximo)  # Limitar entre min y max

        elif distribucion == "triangular":
            c = (moda - minimo) / (maximo - minimo)
            muestras = triang.rvs(c, loc=minimo, scale=(maximo - minimo), size=n)

        elif distribucion == "pert":
            # Distribución PERT (modificada beta): más realista para simulaciones
            def pert_random(a, b, c, lamb=4, size=1):
                alpha = 1 + lamb * (c - a) / (b - a)
                beta_ = 1 + lamb * (b - c) / (b - a)
                samples = beta.rvs(alpha, beta_, size=size)
                return a + samples * (b - a)
            muestras = pert_random(minimo, maximo, moda, size=n)

        df = pd.DataFrame({nombre: muestras})
        tablas[nombre] = df
        distribuciones[nombre] = distribucion

    return tablas, distribuciones
```

```python
def simular_objetivo_montecarlo(celda_objetivo, mapa, tablas_simuladas):
    """
    Evalúa la celda objetivo múltiples veces usando combinaciones de valores simulados de variables raíz.

    Args:
        celda_objetivo (str): Celda a evaluar, por ejemplo 'Hoja!F50'.
        mapa (dict): Diccionario de fórmulas tipo {'Hoja!Celda': lambda r: ...}.
        tablas_simuladas (dict): Diccionario con DataFrames de valores simulados por celda raíz.

    Returns:
        pd.Series: Serie con los valores simulados de la celda objetivo.
    """
    # Unimos todas las tablas simuladas por columnas
    df_simulaciones = pd.concat(tablas_simuladas.values(), axis=1)
    resultados = []

    # Evaluador de celdas usando las fórmulas
    def evaluar(celda, r, mapa, cache={}):
        if celda in cache:
            return cache[celda]
        if celda in r:
            cache[celda] = r[celda]
            return r[celda]
        elif celda in mapa:
            valor = mapa[celda](lambda c: evaluar(c, r, mapa, cache))
            cache[celda] = valor
            return valor
        else:
            raise ValueError(f"No se puede evaluar la celda: {celda}")

    # Iteramos sobre cada fila de simulación
    for _, fila in df_simulaciones.iterrows():
        r_simulado = fila.to_dict()
        try:
            resultado = evaluar(celda_objetivo, r_simulado, mapa, cache={})
        except Exception as e:
            resultado = None  # O np.nan si prefieres
        resultados.append(resultado)

    return pd.Series(resultados)
```

```python
def reconstruir_camino_formula(celda_raiz, celda_objetivo, mapa):
    """
    Reconstruye el camino de dependencias desde una celda raíz hasta una celda objetivo.

    Args:
        celda_raiz (str): Nombre de la celda raíz, por ejemplo 'Costos!F41'.
        celda_objetivo (str): Celda objetivo que se desea alcanzar, por ejemplo 'Finanzas!C50'.
        mapa (dict): Diccionario con fórmulas tipo {'Hoja!Celda': lambda r: ...}.

    Returns:
        list: Lista ordenada de celdas (con sus fórmulas) desde la raíz hasta la celda objetivo.
    """
    visitados = set()
    camino = []

    def dfs(celda_actual, camino_actual):
        if celda_actual in visitados:
            return False
        visitados.add(celda_actual)

        if celda_actual == celda_raiz:
            camino_actual.append(celda_actual)
            return True

        if celda_actual not in mapa:
            return False

        # Obtener las celdas que esta fórmula usa
        try:
            usados = []

            def rastreador(celda_usada):
                usados.append(celda_usada)
                return 0  # Valor ficticio

            mapa[celda_actual](rastreador)
        except Exception:
            return False

        for celda_precedente in usados:
            if dfs(celda_precedente, camino_actual):
                camino_actual.append(celda_actual)
                return True

        return False

    if dfs(celda_objetivo, camino):
        camino.reverse()  # Para ir desde la raíz hasta el objetivo
        return [(celda, mapa[celda]) for celda in camino if celda in mapa]
    else:
        return []
```

```python
def NPV(rate, values):
    return sum(v / (1 + rate) ** i for i, v in enumerate(values, start=1))
```

```python
import re
from openpyxl.utils import column_index_from_string

def excel_to_python(formula_excel):
    """
    Convierte una fórmula de Excel que contiene NPV a código Python.
    - Maneja correctamente rangos como G120:V120.
    - No duplica r('...') en las referencias.
    """
    # Primero, reemplazamos referencias simples (E120, F120) con r('E120'), etc.
    # Pero evitamos hacerlo dentro de NPV(...)
    formula_python = formula_excel
    
    # Extraer NPV y reemplazarlo
    npv_pattern = re.compile(r'NPV\(([^,]+),\s*([A-Z]+\d+):([A-Z]+\d+)\)')
    match = npv_pattern.search(formula_python)
    
    if match:
        tasa = match.group(1)
        celda_inicio = match.group(2)
        celda_fin = match.group(3)
        
        # Extraer fila y columna
        col_inicio = re.match(r'([A-Z]+)(\d+)', celda_inicio).group(1)
        fila = re.match(r'([A-Z]+)(\d+)', celda_inicio).group(2)
        col_fin = re.match(r'([A-Z]+)(\d+)', celda_fin).group(1)
        
        # Generar lista de celdas en el rango (G120, H120, ..., V120)
        celdas = []
        col_num_inicio = column_index_from_string(col_inicio)
        col_num_fin = column_index_from_string(col_fin)
        
        for col_num in range(col_num_inicio, col_num_fin + 1):
            col_letra = column_index_to_string(col_num)  # Función auxiliar
            celdas.append(f"r('{col_letra}{fila}')")
        
        # Reemplazar solo el NPV en la fórmula
        npv_reemplazo = f"NPV({tasa}, [{', '.join(celdas)}])"
        
        formula_python =  npv_reemplazo + formula_python[match.end():]
        print(formula_python)
    
    
    
    return formula_python

# Función auxiliar para convertir número de columna a letra (ej. 7 -> 'G')
def column_index_to_string(col_num):
    letra = ''
    while col_num > 0:
        col_num, resto = divmod(col_num - 1, 26)
        letra = chr(65 + resto) + letra
    return letra
```

```python
import re

def proteger_r_calls(formula):
    """
    Reemplaza temporalmente r('...') por __PROTECTED_n__ para evitar reemplazos dentro.
    """
    protected = {}
    def replacer(match):
        key = f"__PROTECTED_{len(protected)}__"
        protected[key] = match.group(0)
        return key

    nueva_formula = re.sub(r"r\('[^']+'\)", replacer, formula)
    return nueva_formula, protected

def restaurar_r_calls(formula, protected):
    """
    Restaura los r('...') originales en la fórmula.
    """
    for key, value in protected.items():
        formula = formula.replace(key, value)
    return formula

def insertar_r_en_referencias(formula):
    # 1. Proteger r(...) ya existentes
    formula_protegida, protegidos = proteger_r_calls(formula)

    # 2. Reemplazar referencias simples que no están protegidas
    def reemplazo(match):
        ref = match.group(0)
        return f"r('{ref}')"
    
    formula_protegida = re.sub(r"\b([A-Z]{1,3}[0-9]{1,7})\b", reemplazo, formula_protegida)

    # 3. Restaurar lo protegido
    formula_final = restaurar_r_calls(formula_protegida, protegidos)
    return formula_final
```

```python
import re
from collections import defaultdict

class ExcelModel:
    def __init__(self):
        self.cell_map = {}  # Diccionario de celdas: {cell_ref: {'formula': str, 'value': any, 'dependencies': set}}
        self.dependency_graph = defaultdict(set)  # Grafo: {celda: {dependencias}}
        self.reverse_graph = defaultdict(set)     # Grafo inverso: {celda: {dependientes}}
        self.roots = set()                       # Raíces críticas identificadas

    def add_cell(self, cell_ref, formula=None, value=None):
        """Registra una celda en el modelo con validación"""
        if not cell_ref:
            raise ValueError("La referencia de celda no puede estar vacía")
        
        # Validar que solo uno de los dos (formula o value) esté definido
        if formula is not None and value is not None:
            raise ValueError("Una celda no puede tener fórmula y valor fijo simultáneamente")
        
        self.cell_map[cell_ref] = {
            'formula': formula if isinstance(formula, str) else None,
            'value': value,
            'dependencies': self._extract_dependencies(formula) if formula else set()
        }
        
        # Actualizar grafos de dependencias
        for dep in self.cell_map[cell_ref]['dependencies']:
            self.dependency_graph[cell_ref].add(dep)
            self.reverse_graph[dep].add(cell_ref)
            
    def _extract_dependencies(self, formula):
        """Extrae dependencias de una fórmula de Excel de manera segura"""
        if not formula or not isinstance(formula, str):
            return set()
        
        try:
            # Patrón para referencias de celdas (A1, $B$2, Hoja1!C3, etc.)
            pattern = r"(?:'?([\w\s]+)'?\!)?\$?[A-Z]+\$?\d+"
            matches = re.findall(pattern, formula)
            # Filtrar matches vacíos y normalizar formato (quitar comillas innecesarias)
            return {f"'{hoja}'!{celda}" if hoja else celda 
                    for hoja, celda in matches if celda}
        except Exception:
            return set()
    
    def build_dependency_tree(self, target_cell):
        """Construye el árbol de dependencias completo desde la celda objetivo"""
        visited = set()
        self._find_dependencies(target_cell, visited)
        
        # Identificar raíces (celdas sin dependencias)
        self.roots = {cell for cell in visited if not self.cell_map.get(cell, {}).get('dependencies')}
    
    def _find_dependencies(self, cell, visited):
        """Función recursiva para construir el árbol de dependencias"""
        if cell in visited or cell not in self.cell_map:
            return
            
        visited.add(cell)
        
        for dep in self.cell_map[cell]['dependencies']:
            self._find_dependencies(dep, visited)
    
    def evaluate_cell(self, cell, modified_values=None):
        """Evalúa una celda y sus dependencias recursivamente"""
        if modified_values is None:
            modified_values = {}
            
        # Si el valor fue modificado (raíz crítica)
        if cell in modified_values:
            return modified_values[cell]
            
        # Si es una celda base sin fórmula
        cell_data = self.cell_map.get(cell, {})
        if not cell_data.get('formula'):
            return cell_data.get('value')
            
        # Evaluar dependencias primero
        dep_values = {}
        for dep in cell_data['dependencies']:
            dep_values[dep] = self.evaluate_cell(dep, modified_values)
            
        # Evaluar fórmula (requiere implementación específica)
        return self._evaluate_formula(cell_data['formula'], dep_values)
    
    def _evaluate_formula(self, formula, dep_values):
        """Evalúa una fórmula usando los valores de dependencias"""
        # Implementación simplificada (requeriría un parser completo)
        # Para NPV usaríamos nuestra función previamente definida
        if formula.startswith("NPV"):
            return self._evaluate_npv(formula, dep_values)
        # ... otras funciones
        
        # Implementación genérica para fórmulas simples
        try:
            # Reemplazar referencias por valores
            for ref, val in dep_values.items():
                formula = formula.replace(ref, str(val) if val is not None else "0")
            return eval(formula[1:])  # Eliminar "=" inicial
        except:
            return None
    
    def _evaluate_npv(self, npv_formula, dep_values):
        """Evalúa una fórmula NPV específica"""
        # Extraer argumentos usando regex
        match = re.match(r"NPV\(([^,]+),\s*([^)]+)\)", npv_formula)
        if not match:
            return None
            
        rate = float(match.group(1))
        range_ref = match.group(2)
        
        # Obtener valores del rango
        cell_range = self._parse_range(range_ref)
        cashflows = [dep_values.get(cell, 0) for cell in cell_range]
        
        # Usar nuestra función NPV
        return NPV(rate, cashflows)
    
    def sensitivity_analysis(self, target_cell, roots=None):
        """Interfaz para análisis de sensibilidad"""
        if not roots:
            roots = self.roots
            
        print(f"\n=== Análisis de Sensibilidad para {target_cell} ===")
        print(f"Raíces críticas: {', '.join(roots)}")
        
        modified_values = {root: self.cell_map.get(root, {}).get('value') for root in roots}
        
        while True:
            # Mostrar valores actuales
            print("\nValores actuales de raíces:")
            for root in roots:
                print(f"  {root}: {modified_values[root]}")
            
            # Calcular y mostrar resultado objetivo
            result = self.evaluate_cell(target_cell, modified_values)
            print(f"\nResultado objetivo ({target_cell}): {result}")
            
            # Interfaz de modificación
            root_to_modify = input("\nCelda a modificar (o 'exit' para salir): ").strip()
            if root_to_modify.lower() == 'exit':
                break
                
            if root_to_modify not in roots:
                print("¡No es una raíz crítica!")
                continue
                
            try:
                new_value = float(input(f"Nuevo valor para {root_to_modify}: "))
                modified_values[root_to_modify] = new_value
            except ValueError:
                print("¡Valor inválido! Debe ser un número")
```

```python
def obtener_precedentes_y_formulas(ruta_excel, hoja_inicial, celda_inicial):
    ruta_excel = Path(ruta_excel)
    wb = openpyxl.load_workbook(ruta_excel, data_only=False)
    visitados = set()
    datos_celdas = {}  # { "Hoja!A1": {"formula": "=B1+C1", "value": None}, ... }

    def analizar(hoja_nombre, celda_ref):
        clave = f"{hoja_nombre}!{celda_ref}"
        if clave in visitados:
            return
        visitados.add(clave)

        if hoja_nombre not in wb.sheetnames:
            return

        hoja = wb[hoja_nombre]
        celda = hoja[celda_ref]
        valor = celda.value

        datos_celdas[clave] = {
            "formula": str(valor) if isinstance(valor, str) and valor.startswith('=') else None,
            "value": celda.value if not isinstance(valor, str) or not valor.startswith('=') else None
        }

        if isinstance(valor, str) and valor.startswith('='):
            patron = r"(?:'([^']+)'!)?(\$?[A-Z]{1,3}\$?[0-9]{1,7})"
            referencias = re.findall(patron, valor)
            for hoja_ref, celda_obj in referencias:
                hoja_destino = hoja_ref if hoja_ref else hoja_nombre
                celda_destino = celda_obj.upper()
                analizar(hoja_destino, celda_destino)

    analizar(hoja_inicial, celda_inicial.upper())
    del datos_celdas[f"{hoja_inicial}!{celda_inicial.upper()}"]  # Excluir celda objetivo
    return datos_celdas
```

```python
def cargar_desde_excel(ruta_excel, celda_objetivo):
    hoja, celda = celda_objetivo.split("!")
    datos = obtener_precedentes_y_formulas(ruta_excel, hoja, celda)
    print(datos)
    model = ExcelModel()
    for ref, datos_celda in datos.items():
        model.add_cell(ref, formula=datos_celda["formula"], value=datos_celda["value"])
    
    model.add_cell(celda_objetivo, formula=celda_objetivo )  # Agregar la celda objetivo manualmente
    return model
```

```python
def generar_formulas_por_celda(sheet):
    """
    Genera un diccionario con las fórmulas traducidas a Python por celda.

    Args:
        sheet: hoja de cálculo (openpyxl.worksheet, por ejemplo)

    Returns:
        dict: { "E123": "NPV(...)", ... }
    """
    formulas_por_celda = {}

    for row in sheet.iter_rows():
        for cell in row:
            if cell.data_type == "f" and cell.value is not None:
                celda_ref = f"{cell.column_letter}{cell.row}"
                try:
                    formula_excel = cell.value  # fórmula en texto
                    formula_python = traducir_formula_a_python(celda_ref, formula_excel)
                    formulas_por_celda[celda_ref] = formula_python
                except Exception as e:
                    print(f"[ERROR] al traducir fórmula de {celda_ref}: {e}")

    return formulas_por_celda
```

```python
def print_dependency_tree(model, cell, indent=0, max_depth=10, visited=None):
    """Versión mejorada que maneja ciclos y profundidad"""
    if visited is None:
        visited = set()
    
    if cell in visited or indent > max_depth:
        return
        
    visited.add(cell)
    
    # Obtener información de la celda
    cell_data = model.cell_map.get(cell, {})
    formula = cell_data.get('formula')
    value = cell_data.get('value')
    
    # Determinar tipo de celda
    cell_type = "📌 Raíz" if not formula and value is not None else \
               "📝 Fórmula" if formula else \
               "🔢 Valor" if value is not None else \
               "❓ Vacía"
    
    # Imprimir información de la celda
    print("  " * indent + f"├─ {cell} ({cell_type})", end="")
    if value is not None:
        print(f" = {value}", end="")
    if formula:
        print(f" Fórmula: {formula[:50]}...", end="")
    print()
    
    # Imprimir dependencias
    for dep in sorted(model.dependency_graph.get(cell, [])):
        print_dependency_tree(model, dep, indent + 1, max_depth, visited)
```
